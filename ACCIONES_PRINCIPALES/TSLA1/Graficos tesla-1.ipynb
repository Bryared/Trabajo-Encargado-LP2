```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

API_KEY = "DTDu0SOfs4dX4sJgaRSc4FAx7KPzHYBt"
symbol = "TSLA"

# Obtener datos mensuales de los ltimos 60 meses
url = f"https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?serietype=line&timeseries=60&apikey={API_KEY}"
response = requests.get(url)
data = response.json()

# Convertir a DataFrame
df = pd.DataFrame(data["historical"])
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

# Grfico de serie de tiempo
plt.figure(figsize=(12, 6))
plt.plot(df["date"], df["close"], marker="o", color="blue")
plt.title("Precio de Cierre Mensual de las Acciones de Tesla (TSLA)", fontsize=14)
plt.xlabel("Fecha")
plt.ylabel("Precio de Cierre (USD)")
plt.grid(True)
plt.show()
```

Variables:
Fecha (date): Variable categrica ordinal, indica el mes de cotizacin.

Precio de cierre (close): Variable continua, indica el valor en USD al finalizar la sesin burstil

```python
# Histograma del precio de cierre
plt.figure(figsize=(10, 5))
plt.hist(df["close"], bins=12, color="green", edgecolor="black")
plt.title("Distribucin de Precios de Cierre de Tesla (ltimos 5 aos)", fontsize=14)
plt.xlabel("Precio de Cierre (USD)")
plt.ylabel("Frecuencia")
plt.grid(axis="y")
plt.show()
```

Representa la frecuencia de los distintos precios de cierre.

Permite ver en qu rango de precios se han concentrado ms frecuentemente las cotizaciones.
